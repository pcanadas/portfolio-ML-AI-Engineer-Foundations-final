### Handling Missing Data
Let's examine the missing data, starting with the column Teacher_Quality

```python
missing_teacher_quality = students[students[['Teacher_Quality']].isnull().any(axis=1)]
display(missing_teacher_quality)
```

```python
students['Teacher_Quality'].value_counts()
```

In this column there are 78 affected rows. This data is categorical, so when counting the values in that column, we can see that one of the values is widely majority.<br>
Since 1.18% of the rows were affected, we chose to replace with the most frequent category.

```python
students.fillna({'Teacher_Quality': 'Medium'}, inplace=True)
```

```python
students['Teacher_Quality'].unique()
```

We continue with the column 'Parental_Education_Level'

```python
missing_parent_edu_level = students[students[['Parental_Education_Level']].isnull().any(axis=1)]
display(missing_parent_edu_level)
```

```python
students['Parental_Education_Level'].value_counts()
```

There are 90 affected rows in this column. As in the previous case, this data is categorical, so when counting the values ​​in this column, we can see that one of the values is widely majority.<br>
Since 1.36% of the rows are affected, we choose to replace with the most frequent category as we did previously.

```python
students.fillna({'Parental_Education_Level': 'High School'}, inplace=True)
```

```python
students['Parental_Education_Level'].unique()
```

Finally, we will focus on the column 'Distance_from_Home'

```python
missing_distance_home = students[students[['Distance_from_Home']].isnull().any(axis=1)]
display(missing_distance_home)
```

```python
students['Distance_from_Home'].value_counts()
```

There are 67 affected rows in this column. As in the previous case, this data is categorical, so when counting the values in this column, we can see that one of the values is widely majority.
Since 1.01% of the rows are affected, we choose to replace with the most frequent category as we did previously.

```python
students.fillna({'Distance_from_Home': 'Near'}, inplace=True)
```

```python
students['Distance_from_Home'].unique()
```

### Visualizing the data
For a better understanding of the data, we can create visualizations using the Matplotlib library. <br>
We are going to focus on the comparisons with 'Exam_Score' because in the topic we are studying, what we are interested in knowing are the things that interfere with achieving better academic results.

```python
fig, ax = plt.subplots()
ax.bar(students['Hours_Studied'], students['Exam_Score'])
plt.title('Number of Hours Studied per Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

With this graph we can visually observe that the students who have studied the most hours are not necessarily those who have had the best results. We also see some conflicting data in which students obtain good results despite not devoting time to studying.

```python
parental = (students['Parental_Involvement'].value_counts() / len(students['Parental_Involvement']))*100
parental
```

```python
plot = students['Parental_Involvement'].value_counts().plot(kind='pie', autopct='%.2f', 
                                            figsize=(6, 6),
                                            title='Students by Parental Involvement in %')
```

```python
import seaborn as sns
plt.title('Relationship between sleep hours and exam score')
sns.boxplot(data=students, x='Sleep_Hours', y='Exam_Score', hue='Sleep_Hours')
```

This graph is surprising, as it suggests that the average grades obtained are not affected by the hours of nighttime rest. To check that the graph actually shows us the correct data, we calculate the averages of the exam grades according to the hours of sleep:

```python
students_sleeping = students.groupby(['Sleep_Hours'])
students_sleeping['Exam_Score'].mean()
```

```python
plt.title('Influence of attendance on academic results')
sns.scatterplot(data=students, x='Attendance', y='Exam_Score', color='purple')
```

This scatter plot shows the relationship between attendance and academic results. There seems to be a correlation between the two variables: as attendance increases, so do good academic results. While there are also some outliers in the graph, which fall outside the trend, these are at the high values of academic results, so we can assume that there are other factors that influence these students to finally reach these results.<br>
Overall, the graph suggests that high attendance is a significant factor in determining good academic results.

```python
sns.jointplot(data=students, x='Physical_Activity', y='Exam_Score', hue="Exam_Score")
plt.suptitle('Influence of Physical Activity on academic results', y=1.02)
```

With this graph we can see that a moderate physical activity practice favors the improvement of the exam scores
